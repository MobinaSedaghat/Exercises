```python
import math
import numpy as np
import scipy.integrate as spi
import random
f = lambda x: x**3 - 2
intRange0= 0
intRange1 = 2
a = f(intRange1) + 2 #Random numbers range [0,a]
```

```python
fun =[]
hits   = []
misses = []
max = 0
def generate_points(n):

    for point in range(n):
        x,y = (random.uniform(0,a),   random.uniform(0,a))

        if f(x)> max and intRange0 <= x <= intRange1:
          max = f(x)



        if  y <= f(x):
          hits.append((x,y))
        elif x<intRange0 or x>intRange1 or  y > f(x):
          misses.append((x,y))

    return hits, misses


# for a in range(1,100):

#   hits, misses = generate_points(1000)
#   my_result = (intRange1 - intRange0)*a*(len(hits)/(len(hits)+ len(misses)))
#   result = spi.quad(f, intRange0, intRange1)
#   fun.append(float(result[0]) / float(my_result))
#   plt.plot( list(range(1, 100)),fun)

hits, misses = generate_points(1000)
my_result = (intRange1 - intRange0)*a*(len(hits)/(len(hits)+ len(misses)))
result = spi.quad(f, intRange0, intRange1)

print("Hit or Miss: ", my_result)
print("Integral: ", result)
```
